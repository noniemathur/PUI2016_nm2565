```python
from __future__ import print_function

import numpy as np
import pylab as pl
import matplotlib.pyplot as pt
import plotly.plotly as py
from scipy.optimize import curve_fit, minimize
    
%pylab inline
```

```python
dist = ['chisq']
```

```python
my_sample_size = (1500/np.array(range(1, 101)))
my_sample_size
```

```python
assigned_mean = 50
```

```python
assigned_mean
```

```python
df = assigned_mean
```

```python
dic = {}
dic['chisq'] = np.random.chisquare(df, size = 100)
pl.hist(dic['chisq'], bins = 30)
pl.ylabel('N')
pl.xlabel('X')
```

```python
#gaussian = np.random.randn(100)
#plt.hist(gaussian)
#plt.title("Gaussian Histogram")
#plt.xlabel('X')
#plt.ylabel('N')

#fig = plt.gcf()
```

```python
print (dic['chisq'].mean())
```

```python
dic['chisq'] = {} 

for n in my_sample_size:
    dic['chisq'][n] = np.random.chisquare(df, size = n)
    dic['chisq']['means'] = {}
chi_fig = pl.figure(figsize=(10,6)).add_subplot(111)
for nn in dic['chisq']:
    if not type(nn) == str:
        dic['chisq']['means'][nn] = dic['chisq'][nn].mean()
       
        chi_fig.plot(nn, dic['chisq']['means'][nn], 'o')
        chi_fig.set_xlabel('Sample Size', fontsize=18)
        chi_fig.set_ylabel('Sample Mean', fontsize=18)
        chi_fig.set_title('Chi squared', fontsize=18)
        chi_fig.plot([min(my_sample_size), max(my_sample_size)], [df, df], 'k')
```
